# Gráfica f(x)=$e^{\frac{x}{2}} \sin\left(\frac{x}{3}\right)$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,400)
```

```python
# Funciones.
def f(x):
    return np.exp(x/2)*np.sin(x/3)
def g(x):
    return (1/3)*x
def h(x):
    return (((1/3)*x)+((1/6)*(x**2)))
def j(x):
    return (((1/3)*x)+((1/6)*(x**2))+((23/648)*(x**3)))
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,400)
```

```python
# Graficar  funciones.
plt.figure(figsize=(8,6))
plt.plot(x, f(x),label=r"$e^{\frac{x}{2}} \sin\left(\frac{x}{3}\right)$")
plt.plot(x, g(x),label="${1/3x}$")
plt.plot(x, h(x),label="${1/3x+1/6x^2}$")
plt.plot(x, j(x),label="${1/3x+1/6x^2+23/648x^3}$")
plt.axhline(0, color="pink")
plt.axvline(0, color="pink")
plt.xlim(-10, 10)
plt.ylim(-30, 30)
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid() #Color de malla
plt.title("f(x)=" r"$e^{\frac{x}{2}} \sin\left(\frac{x}{3}\right)$") #Agregar titulo
plt.xlabel("Eje x") #Rotulos
plt.ylabel("Eje y") #Rotulos
plt.legend()
plt.show()
```
