# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-5, 5,100)
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return x - 3**(-x)
def g(x):
    return x
def h(x):
    return -3**(-x)
```

```python
f(4)
```

```python
# Graficar  funciones.
plt.figure(figsize=(10,10))
#Al restar la funci칩n g(x) a la funci칩n h(x) nos da como resultado f(x)
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.axhline(0, color="pink")
plt.axvline(0, color="pink")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
#plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.grid() #Color de malla
plt.title("Gr치fica $f(x)=x-3^{-x}$") #Agregar titulo
plt.xlabel("Eje x") #Rotulos
plt.ylabel("Eje y") #Rotulos
plt.show()
```
