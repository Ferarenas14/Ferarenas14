# Gráfica $f(x)=\sqrt{x+1}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-5, 5,100)
```

```python
# Función cuadrática.
def f(x):
    return (x+1)**(1/2)
def g(x):
    return (1+(1/2)*x)
def h(x):
    return (1+((1/2)*x)-(1/8)*x**2)
def j(x):
    return (1+((1/2)*x)-((1/8)*(x**2))+((1/16)*(x**3)))
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-1, 10,100)
```

```python
# Graficar  funciones.
plt.plot(x, f(x),label="$\sqrt{x+1}$")
plt.plot(x, g(x),label="${1+1/2x}$")
plt.plot(x, h(x),label="${1+1/2x-1/8x^2}$")
plt.plot(x, j(x),label="${1+1/2x-1/8x^2+1/16x^3}$")
plt.axhline(0, color="pink")
plt.axvline(0, color="pink")
plt.xlim(-1, 10)
plt.ylim(-1, 5)
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid() #Color de malla
plt.title("Gráfica $f(x)=\sqrt{x+1}$") #Agregar titulo
plt.xlabel("Eje x") #Rotulos
plt.ylabel("Eje y") #Rotulos
plt.legend()
plt.show()
```
