#Polinomio interpolante de Newton

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos de la tabla
F=np.array([-5,1,9,25,55,105])
x=np.array([0,1,2,3,4,5])
```

```python
# Tabla de diferencias
def diferencias_divididas(x, F):
    n = len(F) #cantidad de datos
    coef = np.copy(F)  # Inicializamos con los valores F
    for j in range(1, n): #veces que se va a dividir la diferencia
        for i in range(n - 1, j - 1, -1):
            coef[i] = (coef[i] - coef[i - 1]) / (x[i] - x[i - j]) #calcula cada diferencia dividida y actualiza el valor
    return coef
```
